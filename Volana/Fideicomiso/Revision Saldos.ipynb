```python
# Data histórica de saldos
data_saldos = pd.read_csv("./JERS/data/2024-06-11/SLD_2024-06-11.txt", sep = '|',names = ['contrato','N', 'Saldo'])
data_saldos.head()
```

```python
# Datos de conciliación
data_saldos_conciliar = pd.read_excel("./CONCILIACION VOLANA SALDOS.xlsx", header = 1)
data_saldos_conciliar.columns = [k.replace(" ","_").lower() for k in data_saldos_conciliar.columns]
data_saldos_conciliar.saldo_volana = data_saldos_conciliar.saldo_volana.astype(str)
```

```python
# Filtrando aquellos que no Linq comenta no han sido reportados
data_no_aparece = data_saldos_conciliar.query("saldo_volana.str.contains('NO')")
data_no_aparece.head()
```

```python
# Uniendo los df para saber si han sido reportados o no
data_no_aparece.merge(data_saldos, on = 'contrato', how = 'left')#.query("Saldo.isna()")
```
